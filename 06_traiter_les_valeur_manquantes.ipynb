```python
#la mÃ©thode fillna, qui va nous permettre de remplir certaines informations avec des valeurs.
df["tax"].fillna(0)
```

```python
#Le parametre "method" qui nous permet de spÃ©cifier diffÃ©rentes faÃ§ons de remplir ces donnÃ©es. 
#Par exemple, on peut utiliser "bfill", qui va utiliser les valeurs autour pour remplir notre colonne. 

df["tax"].fillna(method="bfill")
```

```python
#La methode "dropna", qui, elle, va nous permettre d'enlever automatiquement ces valeurs nulles. 
#Et on va lui spÃ©cifier dans le paramÃ¨tre subset les colonnes que l'on souhaite cibler.

df.dropna()
```

```python
# assez agressif. 
# Si on veut, on peut cibler une colonne prÃ©cise, ou plusieurs, 
# en mettant subset, et Ã  l'intÃ©rieur d'une liste, 
# on va mettre les colonnes que l'on souhaite cibler.
df.dropna(subset=["tax"])
```
