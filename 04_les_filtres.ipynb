```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.gender == "Male"
```

```python
# Filtre pour affiche que ceux dont "gender" est "Female" 
female_filter = df.gender == "Female"
df[female_filter]
```

```python
# filtre pour recuperer ceux qui dont "country" est "France", "Canada" en utilisant isin()
contry_filtrer = df.country.isin(["Canada", "France"])
df[contry_filtrer]
```

```python
#Afficher que ceux dont "price_paid" est > 5
df[df.price_paid > 5]
```

```python
#normale
#ceux qu'on peut faire

#on fait une copie de notre dataframe
df_test = df.copy()

#on remplace le symbole $ par ""
df_test.price_paid = df_test.price_paid.apply(lambda x: x.replace("$", ""))

#puis on convertie en float
df_test.price_paid = df_test.price_paid.astype(float)

df_test[df_test["price_paid"] >= 5.0]
```
